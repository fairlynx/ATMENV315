# Python code solution for Confidence interval calculation 

## Example from the textbook Practical Statistics (Ch2. p.18)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats.distributions import norm , t
```

```python
# data
x= np.array([8,12,10,7,7,11,8])
n= np.size(x)
```

```python
m=np.mean(x)
s=np.std(x)

print ("mean and stddev of our data: %8.3f %8.3f" % (m,s))
```

```python
se= s/np.sqrt(n)
print ("standard error of the mean %8.3f" % (se))
```

```python
plt.figure(figsize=[10,8])


# define our t-distrbution

df=n-1
xspace=np.arange(-5,5.1,0.1)
f=t.pdf(xspace,df=df)
plt.plot(xspace,f,label='t-dist')
#plt.plot(xspace,norm.pdf(xspace),label='normal distr.')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[10,8])

df=n-1
xspace=np.arange(-5,5.1,0.1)
F=t.cdf(xspace,df)
plt.plot(xspace,F,label="CDF t-distr.")
plt.hlines(0,-5,5,linestyle='dotted')
plt.legend()
plt.show()
```

```python
t.cdf([-2.5],df)
```

```python
t.ppf([0.025,0.975],df)
```

```python
# 95% confidence interval

ci=se*t.ppf([0.025,0.975],df)
print ("95 percent confidence interval for the mean %6.3f: %6.3f %6.3f" % (m, m+ci[0], m+ci[1]))
```
