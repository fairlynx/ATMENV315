# Fitting grades example

In this notebook we use the two data arrays x and y that contain the sample data:

* x contains the third exam score
* y contains the final exam score 

```python
%matplotlib inline
```

```python
import matplotlib.pylab as plt
import numpy as np
from scipy import stats

x=np.array([ 65, 67, 71, 71, 66, 75, 67, 70, 71, 69, 69])
y=np.array([175,133,185,163,126,198,153,163,159,151,159])
```

## Linear regression with scipy package
It can be done with scipy's stats.linregress function.

```python
#stats.linregress(x,y)
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
```

```python
print(f"y= {intercept:.2f} + {slope:.2f}x")
print(f"correlation coefficient: {r_value:.2f}")
print(f"p-value: {p_value:.6f}")
```

## Create a scatterplot with the regression line 

```python
# Change the default plot style
# You can see what styles are available with print(plt.style.available)
plt.style.use("ggplot")
plt.figure(figsize=(6,6))
# create scatter plot
plt.scatter(x,y,label='sample data')
plt.xlabel("third exam score",fontsize=14)
plt.ylabel("final exam score",fontsize=14)


# get the fitted line y-values for each x value
yhat=intercept+slope*x
# style '-x' gives line connecting the points and using x as a point marker
plt.plot(x,yhat,'x',markersize=10,color='blue',label='fitted data')

# new plot function: insert an annotated arrow 
plt.annotate("outlier ?",(65.2,174.5),(67,170),\
             fontsize=16,arrowprops={"arrowstyle":"simple","color":"gray"})

# adding a linear line with the function plt.abline
# it needs to arguments, the intercept and the slope
plt.plot(np.sort(x),intercept+ slope*np.sort(x),':',color='purple',label='fitted line') # best is to sort the x data
plt.legend()
plt.show()
```

## Summary

It is quite simple to calculate and plot a linear regression line. This notebook illustrates  
that once you have your paired data samples in numpy arrays, only a few lines of code produce
a decent plot. You can start from this template and be creative and improve the style of the figure.
