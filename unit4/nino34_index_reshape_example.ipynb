```python
import numpy as np
import matplotlib.pyplot as plt

matrix= np.loadtxt("../data/nino34_index.csv",skiprows=8)
# take the data matrix and extract column 0 (years)
# and columns 1-12 (Jan to Dec monthly mean SST anomalies)
year=matrix[:,0]
nino34=matrix[:,1:13]

```

```python
first, last = year[0], year[-1]
nyear=len(year)
print("First and last year in file: "+str(int(first))+" "+str(int(last)))
print("Number of years: "+str(nyear))
```

```python
# We want to see the monthly mean time series!
# For this we have to create a 1-dimensional array
# from the matrix nino34

# print shape and size of nino34

print("shape of nino34 matrix: ")
print(nino34.shape)
print("number of values stored in nino34 matrix: ")
nsize=nino34.shape
print(nsize)





print ("create 1-dim array with same size (length) and plot time series")
# from 2-dim to 1-dim array
nino1d=np.reshape(nino34, 1812 )# or use np.reshape(nino34, -1) see references below
print(nino1d.shape)



# simple time index with np.linspace(start,end, number of values )
time = np.arange(first,last+1,1/12.)
```

```python
plt.figure(figsize=[8,4])
plt.plot(time,nino1d)
plt.xlim([1950,2020])
plt.show()
# compare with NCAR's web page plot
from IPython.display import Image
Image("../data/nino34_noaa_img.gif",width=600) # Image is a function that we call to display the imag
```
