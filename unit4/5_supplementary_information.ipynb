Example for arranging 24 numbers in a 1-dimensional array and then reshaping it to a 2-dimensional array. 
This is the notebook with code supporting the demo with the colored tiles.


```python
import numpy as np
import matplotlib.pyplot as plt
x= np.array([ 1,2,3,4,5,6,1,2,3,4,5,6,1,2,3,4,5,6,1,2,3,4,5,6])

print("shape of array x: {0}".format(x.shape))

i=0

#print the index positions and array values in two lines
line1="index    i: "
line2="value x[i]: "
i=0
while i< np.size(x):
    line1=line1+" %2d " % (i)
    line2=line2+" %2d " % (x[i])
    i=i+1
print(line1)
print(line2)
```

### In the next cell we define a function that we use to plot the array


```python
def show_2d_array(x):
    s=x.shape
    n=s[0] # rows
    m=s[1] # columns
    cm=plt.get_cmap('rainbow')
    fig=plt.figure()
    plt.imshow(x,origin='upper',cmap=cm,alpha=0.4)
    plt.xticks(np.arange(0,m,1))
    plt.yticks(np.arange(0,n,1))
    i=0
    while i< m: # columns
        j=0
        while j<n: # rows:
            plt.text(i,j,"["+str(j)+","+str(i)+"]",fontsize=12,
                     horizontalalignment='center',
                     verticalalignment='center',
                     color="black")
            j=j+1
        i=i+1
    plt.grid()
    plt.colorbar()
    plt.show()
    
```

### Reshape function 
The two reshape examples illustrate how the values from the 1-dimensional array are ordered
inside the rows and columns.

```python
# reshape function applied to turn into 4 x 6 array
x2d=np.reshape(x,newshape=[4,6])
show_2d_array(x2d)
```

The figure above shows a 4 x 6 array.

```python
# reshape function applied to turn into 4 x 6 array
x2d=np.reshape(x,newshape=[6,4])
show_2d_array(x2d)
```

The figure above shows a 6 x 4 array.

### Summary: 
The reshaping of arrays from 1-dimensional into 2-dimensional is possible with _np.reshape()_.
The default method is filling the data in row by row.
