## What are subplots?

Subplots are a function in matplotlib that allows you to make multiple plots in one large plot

### add example image here

```python
### Begin by importing matplotlib
import matplotlib.pyplot as plt

import pandas as pd
import xarray as xr
```

```python
#Data importing
aircraft = xr.open_dataset('/network/rit/lab/minderlab_rit/ms957639/WINTREMIXDATA/aircraft/atmospheric-state_20220307_F07.nc')

time=aircraft['time']
temp=aircraft['Ts']
pres=aircraft['Ps']
dew=aircraft['Td_CM']
```

## Plotting with Matplotlib

### How to start plotting


```python
#Set figure to plot on 
fig, ax = plt.subplots()
```

```python
fig, ax = plt.subplots()

#Plot your data
ax.plot(time,temp, color='red', label='Line Plot')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(time,temp, color='red', label='Line Plot')

##Clean up plot
#add X & Y labels
ax.set_xlabel('Time (UTC)')
ax.set_ylabel('Temperature (degC)')

#add plot title
ax.set_title('Temperature Timeseries')

#Rotate x axis ticks to not overlap
ax.tick_params(axis='x', labelrotation = 20)

#Show legend
ax.legend()
plt.show()
```

## What can we make with Matplotlib?

Well, a lot different plots and as scientists studying the environment it is crucial we can make informative and aesthetically pleasing plots. Here we will explore the capabilities of matplotlib

### Line Plots & Timeseries

```python
fig, ax = plt.subplots()
ax.plot(time,temp, color='red', label='Line Plot')
ax.set_xlabel('Time (UTC)')
ax.set_ylabel('Temperature (degC)')
ax.set_title('Temperature Timeseries')
ax.tick_params(axis='x', labelrotation = 20)
ax.legend()
plt.show()
```

### Scatter Plot

```python
fig, ax = plt.subplots()
ax.scatter(time,temp, color='maroon', marker = '*', label='Scatter Plot')
ax.set_xlabel('Time (UTC)')
ax.set_ylabel('Temperature (degC)')
ax.set_title('Temperature Timeseries')
ax.tick_params(axis='x', labelrotation = 20)
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(time,dew, color='g', marker = 'o', label='Scatter Plot')
ax.set_xlabel('Time (UTC)')
ax.set_ylabel('Temperature (degC)')
ax.set_title('Dew Point Timeseries')
ax.tick_params(axis='x', labelrotation = 20)
ax.legend()
plt.show()
```

### Box Plot

```python
## Box plot
fig, ax = plt.subplots()
#Add data
ax.boxplot(temp)
#Set labels
ax.set_ylabel('Temperature (degC)')
ax.set_title('Temperature Box Plot')
plt.show()
```

## Subplots
### and two different ways to do them

Subplots are useful when you would like to make more than one plot within a figure. The way to use them is similar to one plot, let's walk through it!

### If you have a few plots

```python
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(7,7))
#fig, ax = plt.subplots(rows, cols)
```

```python
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(10,10), sharey=True)

#Plotting on axis
ax0.plot(time, temp, color='red')
ax1.plot(time, dew, color ='green')

#Rotate x labels
ax0.tick_params(axis='x', labelrotation = 20)
ax1.tick_params(axis='x', labelrotation = 20)

ax0.set_xlabel('Time (UTC)')
ax0.set_ylabel('Temperature (degC)')
ax0.set_title('Temperature Timeseries')

ax1.set_xlabel('Time (UTC)')
ax1.set_ylabel('Temperature (degC)')
ax1.set_title('Dew Point Timeseries')

plt.show()
```

### If you have many axes

```python
fig, axs = plt.subplots(2, figsize=(7,7))
```

```python
fig, axs = plt.subplots(2, figsize=(10,10), sharex=True)
axs[0].plot(time, temp, color='red')
axs[1].plot(time, dew, color='green')

axs[0].tick_params(axis='x', labelrotation = 20)
axs[1].tick_params(axis='x', labelrotation = 20)

axs[0].set_ylabel('Temperature (degC)')
axs[0].set_title('Temperature Timeseries')

axs[1].set_xlabel('Time (UTC)')
axs[1].set_ylabel('Temperature (degC)')
axs[1].set_title('Dew Point Timeseries') 

plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12,12))
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12,12))

ax[0,0].plot(time,temp, color='red', label='Line Plot')
ax[0,1].scatter(time,temp, color='maroon', marker='*', label='Scatter Plot')
ax[1,0].scatter(time,dew, color='g', marker='o',label='Scatter Plot')
ax[1,1].boxplot(temp)

ax[0,0].set_xlabel('Time (UTC)')
ax[0,0].set_ylabel('Temperature (degC)')
ax[0,0].set_title('Temperature Timeseries')
ax[0,0].tick_params(axis='x', labelrotation = 20)


ax[0,1].set_xlabel('Time (UTC)')
ax[0,1].set_ylabel('Temperature (degC)')
ax[0,1].set_title('Temperature Timeseries')
ax[0,1].tick_params(axis='x', labelrotation = 20)

ax[1,0].set_xlabel('Time (UTC)')
ax[1,0].set_ylabel('Temperature (degC)')
ax[1,0].set_title('Dewpoint Timeseries')
ax[1,0].tick_params(axis='x', labelrotation = 20)

ax[1,1].set_ylabel('Temperature (degC)')
ax[1,1].set_title('Temperature Box Plot')

plt.show()
```

## Important take away

The syntax is not consistent between how we have been plotting in class and using subplots. Many function names will changes so it is important to be aware of such changes.

The following change in the function calls should be remembered when using subplot
axes. If you have one of your subplot axes in variable ___ax___ then use the following function calls instead of the plt-function calls ($\Rightarrow$ indicates which function call to use instead!):

- `plt.xlabel()` $\Rightarrow$ `ax.set_xlabel()`

- `plt.ylabel()` $\Rightarrow$ `ax.set_ylabel()`

- `plt.xlim()` $\Rightarrow$ `ax.set_xlim()`
- `plt.ylim()` $\Rightarrow$ `ax.set_ylim()`

- `plt.title()` $\Rightarrow$ `ax.set_title()`
